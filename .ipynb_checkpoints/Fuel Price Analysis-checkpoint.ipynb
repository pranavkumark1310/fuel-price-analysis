## Import packages

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read dataset
df = pd.read_csv('petrol.csv')
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
#check for duplicate values in the dataset.
df.duplicated().sum()
```

```python
#check for any null value in the dataset.
df.isnull().sum()
```

```python
#find duplicate row in dataset
duplicate_row = df[df.duplicated()]
print(duplicate_row)
```

```python
#convert date from string to datetime format
df['date'] = pd.to_datetime(df['date'])
```

```python
#view revised dataset changes
df.info()
```

## Exploratory Data Analysis

### Scatterplot analysis - Date vs Rate

```python
a4_dims = (11.7, 8.27)
fig, ax = plt.subplots(figsize=a4_dims)
sns.scatterplot(x="date", y="rate",
                palette="ch:r=-.2,d=.3_r",
                sizes=(7, 8), linewidth=0,
                data=df)
```

### Lineplot analysis - Date vs Rate

```python
a4_dims = (11.7, 8.27)
fig, ax = plt.subplots(figsize=a4_dims)
sns.lineplot(x="date", y="rate",data=df)
```

```python
tamilnadu_data = df[df['state'] == 'Tamil Nadu'] 
print(tamilnadu_data)
```

```python
delhi_data = df[df['state'] == 'Delhi'] 
print(delhi_data)
```

```python
telangana_data = df[df['state'] == 'Telangana'] 
print(telangana_data)
```

```python
karnataka_data = df[df['state'] == 'Karnataka'] 
print(karnataka_data)
```

```python
maharashtra_data = df[df['state'] == 'Maharashtra'] 
print(maharashtra_data)
```

### Tamil Nadu petrol price from 2008 to 2020

```python
x = np.array(tamilnadu_data.date)
y = np.array(tamilnadu_data.rate)
plt.figure(figsize=(12,8))
plt.title('Tamil Nadu Petrol price from 2008 to 2020',fontsize=20)
plt.ylabel('Petrol Rate',fontsize=16)
plt.xlabel('Year',fontsize=16)
plt.plot(x,y)
```
